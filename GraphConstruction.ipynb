```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations
import hypernetx as hnx
import numpy as np
import warnings 
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv("final_dataset.csv", sep=',')
df.head()
```

```python
df=df[['gene', 'disease']]
```

```python
df.describe()
```

```python
H = hnx.Hypergraph(df)
```

```python
H.nodes.items
```

```python
H.edges.items
```

```python
H.incidences.items
```

```python
hnx.info_dict(H)
```

```python
H_node_collapse_counts,eclass_counts = H.collapse_nodes(use_counts=True, return_equivalence_classes=True)
eclass_counts
```

```python
collapsed_result = H.collapse_nodes(return_equivalence_classes=True) #collapsed_result=(H_collapsed, equivalence_classes)
H_collapsed = collapsed_result[0]
node_equiv_classes = {rep: members for rep, members in collapsed_result[1].items() if rep in H_collapsed.nodes()}

new_hyperedges = {}
for hyperedge in H_collapsed.edges:
    nodes = H_collapsed.edges[hyperedge]  
    new_nodes = set()
    for node in nodes:
        rep = None
        for r, members in node_equiv_classes.items():
            if node in members:
                rep = r
                break
        if rep is None:
            rep = node  
        new_nodes.add(rep)
    new_hyperedges[hyperedge] = new_nodes

H_final = hnx.Hypergraph(new_hyperedges)

print("Nodi (malattie, rinominate):", list(H_final.nodes()))
print("Archi (iperarchi, geni, invariati):", list(H_final.edges()))
```

```python
hnx.info_dict(H_final)
```

```python
rows = []
for hyperedge in H_final.edges:
    for node in H_final.edges[hyperedge]:
        rows.append({'gene': hyperedge, 'disease': node})

df_long = pd.DataFrame(rows)
print(df_long)
```

```python
df_long = df_long[~df_long['disease'].str.contains('disease', case=False, na=False)]
```
